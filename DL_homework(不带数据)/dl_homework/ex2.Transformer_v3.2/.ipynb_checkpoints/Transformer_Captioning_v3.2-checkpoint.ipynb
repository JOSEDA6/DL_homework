# Transformer
As you have seen, RNNs are incredibly powerful but often slow to train. Further, RNNs struggle to encode long-range dependencies (though LSTMs are one way of mitigating the issue). In 2017, Vaswani et al introduced the Transformer in their paper ["Attention Is All You Need"](https://arxiv.org/abs/1706.03762) to a) introduce parallelism and b) allow models to learn long-range dependencies. The paper not only led to famous models like BERT and GPT in the natural language processing community, but also an explosion of interest across fields, including vision. While here we introduce the model in the context of image captioning, the idea of attention itself is much more general.



# Transformer: Multi-Headed Attention

### Dot-Product Attention

Recall that attention can be viewed as an operation on a query $q\in\mathbb{R}^d$, a set of value vectors $\{v_1,\dots,v_n\}, v_i\in\mathbb{R}^d$, and a set of key vectors $\{k_1,\dots,k_n\}, k_i \in \mathbb{R}^d$, specified as

\begin{align}
c &= \sum_{i=1}^{n} v_i \alpha_i \\ \alpha_i &= \frac{\exp(k_i^\top q)}{\sum_{j=1}^{n} \exp(k_j^\top q)} \\
\end{align}

where $\alpha_i$ are frequently called the "attention weights", and the output $c\in\mathbb{R}^d$ is a correspondingly weighted average over the value vectors.

### Self-Attention
In Transformers, we perform self-attention, which means that the values, keys and query are derived from the input $X \in \mathbb{R}^{\ell \times d}$, where $\ell$ is our sequence length. Specifically, we learn parameter matrices $V,K,Q \in \mathbb{R}^{d\times d}$ to map our input $X$ as follows:

\begin{align}
v_i = Vx_i\ \ i \in \{1,\dots,\ell\}\\
k_i = Kx_i\ \ i \in \{1,\dots,\ell\}\\
q_i = Qx_i\ \ i \in \{1,\dots,\ell\}
\end{align}

### Multi-Headed Scaled Dot-Product Attention
In the case of multi-headed attention, we learn a parameter matrix for each head, which gives the model more expressivity to attend to different parts of the input. Let $h$ be number of heads, and $Y_i$ be the attention output of head $i$. Thus we learn individual matrices $Q_i$, $K_i$ and $V_i$. To keep our overall computation the same as the single-headed case, we choose $Q_i \in \mathbb{R}^{d\times d/h}$, $K_i \in \mathbb{R}^{d\times d/h}$ and $V_i \in \mathbb{R}^{d\times d/h}$. Adding in a scaling term $\frac{1}{\sqrt{d/h}}$ to our simple dot-product attention above, we have

\begin{equation}
Y_i = \text{softmax}\bigg(\frac{(XQ_i)(XK_i)^\top}{\sqrt{d/h}}\bigg)(XV_i)
\end{equation}

where $Y_i\in\mathbb{R}^{\ell \times d/h}$, where $\ell$ is our sequence length.

In our implementation, we apply dropout to the attention weights (though in practice it could be used at any step):

\begin{equation}
Y_i = \text{dropout}\bigg(\text{softmax}\bigg(\frac{(XQ_i)(XK_i)^\top}{\sqrt{d/h}}\bigg)\bigg)(XV_i)
\end{equation}

Finally, then the output of the self-attention is a linear transformation of the concatenation of the heads:

\begin{equation}
Y = [Y_1;\dots;Y_h]A
\end{equation}

were $A \in\mathbb{R}^{d\times d}$ and $[Y_1;\dots;Y_h]\in\mathbb{R}^{\ell \times d}$.

Implement multi-headed scaled dot-product attention in the `MultiHeadAttention` class in the file `sources/transformer_layers.py`. The code below will check your implementation. The relative error should be less than `e-2`.

# Positional Encoding

While transformers are able to easily attend to any part of their input, the attention mechanism has no concept of token order. However, for many tasks (especially natural language processing), relative token order is very important. To recover this, the authors add a positional encoding to the embeddings of individual word tokens.

Let us define a matrix $P \in \mathbb{R}^{l\times d}$, where $P_{ij} = $

$$
\begin{cases}
\text{sin}\left(i \cdot 10000^{-\frac{j}{d}}\right) & \text{if j is even} \\
\text{cos}\left(i \cdot 10000^{-\frac{(j-1)}{d}}\right) & \text{otherwise} \\
\end{cases}
$$

Rather than directly passing an input $X \in \mathbb{R}^{l\times d}$ to our network, we instead pass $X + P$.

Implement this layer in `PositionalEncoding` in `sources/transformer_layers.py`. Once you are done, run the following to perform a simple test of your implementation. You should see errors on the order of `e-3` or less.

```python
torch.manual_seed(231)

batch_size = 1
sequence_length = 2
embed_dim = 6
data = torch.randn(batch_size, sequence_length, embed_dim)

pos_encoder = PositionalEncoding(embed_dim)
pos_encoder.eval()
output = pos_encoder(data)

expected_pe_output = np.asarray(
    [[[-1.1106,  1.0014,  1.5280, -0.0778, -0.6964,  1.1455],
         [ 0.8125, -0.4303,  0.4981,  0.7320,  1.1380,  1.5331]]]
)

print('pe_output error: ', rel_error(expected_pe_output, output.detach().numpy()))
```

# Transformer for Image Captioning

```python
import time, os, json
import numpy as np
import matplotlib.pyplot as plt
import torch

from sources.gradient_check import eval_numerical_gradient, eval_numerical_gradient_array
from sources.transformer_layers import *
from sources.captioning_solver_transformer import CaptioningSolverTransformer
from sources.transformer import CaptioningTransformer
from sources.coco_utils import load_coco_data, sample_coco_minibatch, decode_captions
from sources.image_utils import image_from_url

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0)  # Set default size of plots.
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%load_ext autoreload
%autoreload 2

def rel_error(x, y):
    """returns relative error"""
    return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))

# -------------------------------
# Load COCO data from disk into a dictionary.
data = load_coco_data(pca_features=True)
print("COCO数据加载完毕，共有%d个键。" % len(data.keys()))
torch.manual_seed(231)

# -------------------------------
# Train model (only final loss and loss curve are shown)
data = load_coco_data(max_train=50)
transformer = CaptioningTransformer(
    word_to_idx=data['word_to_idx'],
    input_dim=data['train_features'].shape[1],
    wordvec_dim=256,
    num_heads=2,
    num_blocks=2,
    max_length=30
)
transformer_solver = CaptioningSolverTransformer(
    transformer, data, idx_to_word=data['idx_to_word'],
    num_epochs=100,
    batch_size=25,
    learning_rate=0.001,
    verbose=False,       # turn off verbose output
    print_every=10,
)
transformer_solver.train()
print('Final loss: ', transformer_solver.loss_history[-1])

plt.plot(transformer_solver.loss_history)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training loss history')
plt.show()

# -------------------------------
# Sample images and display captions
for split in ['train', 'val']:
    minibatch = sample_coco_minibatch(data, split=split, batch_size=2)
    gt_captions, features, urls = minibatch
    gt_captions = decode_captions(gt_captions, data['idx_to_word'])
    sample_captions = transformer.sample(features, max_length=30)
    sample_captions = decode_captions(sample_captions, data['idx_to_word'])
    
    for gt_caption, sample_caption, url in zip(gt_captions, sample_captions, urls):
        try:
            img = image_from_url(url)
        except Exception as e:
            img = None
        if img is None:
            continue
        plt.imshow(img)
        plt.title('%s\n%s\nGT:%s' % (split, sample_caption, gt_caption))
        plt.axis('off')
        plt.show()

# -------------------------------
# ----- Visualize Attention Weights -----
# Updated algorithm for generating attention heatmaps:
# 1. For each selected time step, extract the 1D attention vector (length = L).
# 2. Normalize the attention vector to [0,1] and apply a gamma correction to enhance contrast.
# 3. Compute the outer product of the gamma-corrected vector with itself to obtain a 2D matrix.
# 4. Use bilinear interpolation to resize the heatmap to the image dimensions.
# 5. Multiply it by a scaling factor to adjust the dynamic range.
# 6. Overlay the resulting heatmap on the original image.
# Intermediate parameters are printed for debugging.

import urllib.request
from urllib.error import HTTPError, URLError
from PIL import Image

def safe_image_from_url(url, temp_file_path="temp_image.jpg"):
    """
    Try to download an image from URL with graceful error handling.
    If downloading fails, return None.
    """
    try:
        urllib.request.urlretrieve(url, temp_file_path)
        img = Image.open(temp_file_path)
        return np.array(img)
    except Exception:
        return None
    finally:
        try:
            if os.path.exists(temp_file_path):
                os.remove(temp_file_path)
        except PermissionError:
            pass

def resize_heatmap(heatmap, target_shape):
    """
    Resize the 2D heatmap to "target_shape" (height, width) using bilinear interpolation.
    """
    heatmap_norm = 255 * (heatmap - np.min(heatmap)) / (np.ptp(heatmap) + 1e-6)
    heatmap_img = Image.fromarray(np.uint8(heatmap_norm))
    heatmap_resized = heatmap_img.resize((target_shape[1], target_shape[0]), resample=Image.BILINEAR)
    return np.array(heatmap_resized) / 255.0

def visualize_attention(model, features, idx_to_word, image_url=None, word_step=4, heatmap_alpha=0.85, heatmap_scale=5.0, gamma=2.5):
    """
    Visualize attention distributions in the Transformer model.
    
    Display:
      - Top row: Original image, with titles showing every "word_step"-th generated word.
      - Bottom row: Overlaid attention heatmaps. For each selected time step:
            a. Take the 1D attention vector (length L).
            b. Normalize it to [0,1] and apply gamma correction to enhance contrast.
            c. Compute the outer product (2D heatmap).
            d. Resize the heatmap to match the image dimensions using bilinear interpolation.
            e. Multiply by heatmap_scale to further amplify differences.
            f. Overlay on the image.
            
    Intermediate stats (mean, std, min, max) are printed for debugging purposes.
    
    Parameters:
      word_step: Interval for selecting time steps.
      heatmap_alpha: Transparency for the heatmap overlay.
      heatmap_scale: Scaling factor for the heatmap values.
      gamma: Gamma correction exponent to enhance contrast.
    """
    if not isinstance(features, torch.Tensor):
        features = torch.tensor(features)
    
    model.eval()
    with torch.no_grad():
        sampled_caption = model.sample(features, max_length=20)
        if not isinstance(sampled_caption, torch.Tensor):
            sampled_caption = torch.tensor(sampled_caption, dtype=torch.long, device=features.device)
        _ = model(features, sampled_caption)
        attn_weights = model.transformerdecoder.layers[0].self_attn.attn_weights
        attn = attn_weights[0, 0].cpu().numpy()  # shape: (L, L)
        
        if isinstance(idx_to_word, dict):
            caption_words = [idx_to_word.get(int(idx), "<UNK>") for idx in sampled_caption[0]]
        else:
            caption_words = [idx_to_word[int(idx)] if int(idx) < len(idx_to_word) else "<UNK>" 
                             for idx in sampled_caption[0]]
    
    L = attn.shape[0]
    selected_indices = list(range(0, L, word_step))
    selected_words = [caption_words[i] for i in selected_indices]
    selected_attn = attn[selected_indices, :]  # shape: (num_selected, L)
    
    print("DEBUG: Selected time steps:", selected_indices)
    for i, vec in enumerate(selected_attn):
        print(f"DEBUG: Time {selected_indices[i]} - vector: mean={np.mean(vec):.4f}, std={np.std(vec):.4f}, min={np.min(vec):.4f}, max={np.max(vec):.4f}")
    
    if image_url is not None:
        img = safe_image_from_url(image_url)
        if img is None:
            img = np.ones((224, 224, 3))
    else:
        img = np.ones((224, 224, 3))
    
    num_plots = len(selected_indices)
    fig, axes = plt.subplots(2, num_plots, figsize=(2.5 * num_plots, 8))
    
    # Top row: display the original image with word titles
    for j, word in enumerate(selected_words):
        ax = axes[0, j]
        ax.imshow(img)
        ax.set_title(word)
        ax.axis("off")
    
    # Bottom row: generate and overlay the heatmaps
    for j, idx in enumerate(selected_indices):
        ax = axes[1, j]
        ax.imshow(img)
        attn_vector = selected_attn[j]  # (L,)
        
        # Normalize the attention vector and apply gamma correction
        vec_min, vec_max = np.min(attn_vector), np.max(attn_vector)
        attn_vector_norm = (attn_vector - vec_min) / (np.ptp(attn_vector) + 1e-6)
        attn_vector_gamma = np.power(attn_vector_norm, gamma)
        print(f"DEBUG: Time {idx} - normalized vector: mean={np.mean(attn_vector_norm):.4f}, std={np.std(attn_vector_norm):.4f}")
        print(f"DEBUG: Time {idx} - gamma-corrected vector: mean={np.mean(attn_vector_gamma):.4f}, std={np.std(attn_vector_gamma):.4f}")
        
        # Compute outer product to obtain 2D heatmap
        heatmap_raw = np.outer(attn_vector_gamma, attn_vector_gamma)
        print(f"DEBUG: Time {idx} - raw heatmap: min={np.min(heatmap_raw):.4f}, max={np.max(heatmap_raw):.4f}, mean={np.mean(heatmap_raw):.4f}")
        
        # Scale and resize heatmap
        heatmap_raw *= heatmap_scale
        heatmap_resized = resize_heatmap(heatmap_raw, (img.shape[0], img.shape[1]))
        print(f"DEBUG: Time {idx} - resized heatmap: min={np.min(heatmap_resized):.4f}, max={np.max(heatmap_resized):.4f}, mean={np.mean(heatmap_resized):.4f}")
        
        ax.imshow(heatmap_resized, cmap="jet", alpha=heatmap_alpha)
        ax.set_title("Attn")
        ax.axis("off")
    
    plt.tight_layout()
    plt.show()

# Example usage: visualize attention overlay for a sampled image
minibatch = sample_coco_minibatch(data, split="val", batch_size=1)
gt_captions, features, urls = minibatch
visualize_attention(transformer, features, data["idx_to_word"], urls[0] if urls else None)
```
